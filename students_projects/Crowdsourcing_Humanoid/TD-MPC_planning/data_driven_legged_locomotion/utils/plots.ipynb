```python
%matplotlib tk
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ast
```

```python
data_path = '/home/antonio/uni/tesi/data-driven-legged-locomotion/experiments/h1_walk_2024-07-29_11-21-46/experiment_data.csv'
S = 2
data = pd.read_csv(data_path)

time = data['Time'].to_numpy()

# Get states
States = data['State'].apply(ast.literal_eval).tolist()
States = np.array(States)

# Get services next states
NextStates = []

for i in range(S):
    l = data['Service_{}_NextState'.format(i)].apply(ast.literal_eval).tolist()
    NextStates.append(l)
NextStates = np.array(NextStates)
NextStates.shape # (S, N, 51)
```

### Plot activated services

```python
service_indexes = data['Service_Index'].to_numpy()
plt.figure()
plt.title('Activated Services')
plt.xlabel('Time [s]')
plt.ylabel('Service Index')
#plt.xlim(1.5,2.5)
plt.stem(time,service_indexes, basefmt=' ')
```

### Plot next states

```python
plt.figure()
plt.title('Next states generated by services')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(States[:, 0], States[:, 1], label='Robot Trajectory', color='black')
for i in range(S):
    plt.scatter(NextStates[i, :, 0], NextStates[i, :, 1], label='Service {}'.format(i), marker='x')
plt.legend()
```
