```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111,projection='3d')

# Prepare arrays x, y, z

z = 5
x = np.arange(1,5,1)
y = np.arange(1,5,1)

x,y = np.meshgrid(x,y)

z = x*2
Axes3D.plot_surface(x,y,z,rstride=1, cstride=1, cmap=matplotlib.jet, linewidth=1, antialiased=True)  #这里传入x, y, z的值

plt.show()
```

```python
# 绘制三维曲面
fig = plt.figure()
 
axes3d = Axes3D(fig)
 
#!！面
x = np.arange(1,3,1)
y = np.arange(1,3,1)
 
    
X,Y = np.meshgrid(x,y)
Z = X*0+Y*0+2
 
axes3d.plot_surface(X,Y,Z)
plt.show()
```
