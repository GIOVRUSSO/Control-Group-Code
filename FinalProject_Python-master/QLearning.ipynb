```python
"""
Reinforcement Learning using table lookup Q-learning method.

"""

import numpy as np
import pandas as pd
import time
```

```python
np.random.seed(2)  # reproducible
```

## **Initial all parameters**

```python
N_STATES = 41   # heart rate from 0 - 200
                # in the begin, I only considering the HR, later, when the code can run sccessful, human power should be another aspect state

ACTIONS = ['1', '2', '3', '4', '5']     # available actions, possible throttle out, to simplify the problem, only consider int

# TODO insert 
EPSILON = 0.9   # greedy police

ALPHA = 0.5     # learning rate, step size

GAMMA = 0.9     # discount factor

MAX_EPISODES = 100 # maximum episodes, this number should large than the possible ride minutes

FRESH_TIME = 60    # fresh time for one move, cooresponse to the states update rate
```

## **Build initialzed Q-table**

```python
def build_q_table(n_states, actions):
    table = pd.DataFrame(
        np.zeros((n_states, len(actions))),     # q_table initial values
        columns=actions,    # actions's name, possible int voltage output
        index = [np.arange(0,205,5)], # heart rate (D5)
    )
    #print(table)    # show table
    return table

build_q_table(N_STATES,ACTIONS)  # initial the table
```

## **Define the action policy**
#### e-greedy policy is applied in this case

```python
def choose_action(state, q_table):
    state_actions = q_table.iloc[state, :]
    if (np.random.uniform() > EPSILON) or ((state_actions == 0).all()):  # act non-greedy or state-action have no value
        action_name = np.random.choice(ACTIONS)
    else:   # act greedy
        action_name = state_actions.idxmax()
    return action_name
```

```python
def rl(S):
    # main part of RL loop
    q_table = build_q_table(N_STATES,ACTIONS)
    
    for episode in range(MAX_EPISODES):
        
        step_counter = 0
        
       # beginning position, S= 0 is for simplify the problem, better to  measure the baseline of cyclists
        int S = 0
        
        is_terminated = False
        
        #update_env(S, episode, step_counter+1)
        
        while not is_terminated:

            A = choose_action(S, q_table)
            S_, R = get_env_feedback(S, A)  # take action & get next state and reward
            q_predict = q_table.loc[S, A]
            if S_ != 'terminal':
                q_target = R + GAMMA * q_table.iloc[S_, :].max()   # next state is not terminal
            else:
                q_target = R     # next state is terminal
                is_terminated = True    # terminate this episode

            q_table.loc[S, A] += ALPHA * (q_target - q_predict)  # update
            S = S_  # move to next state

            #update_env(S, episode, step_counter+1)
            step_counter += 1
            
    return q_table
```
